```python
%matplotlib inline
import numpy as np
import sys
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import importlib
import time
import datetime
import segment_cluster as sc
import pandas as pd
import os
import fnmatch

importlib.reload(sc)
sys.stdout.flush()
np.random.seed(0)

k_clusters = [10, 50, 100, 200]
seg_lens = [10, 50, 100, 150, 200]
```

```python
def plot_time_series(training_file, outlier_file):
    fig, axes = plt.subplots(nrows=5, ncols=1)
    for index, time_series in enumerate(training_file[0:5]):
        axes[index].plot(time_series)
        axes[index].plot(outlier_file[index])
        axes[index].get_xaxis().set_visible(False)
    axes[-1].get_xaxis().set_visible(True)
    fig.tight_layout()
    fig.show()
    
    return plt.show()
    
def validation_and_analysis(ordinary_file, outlier_file, k_clusters, seg_lens):
    file_name, validation_results = sc.validate_algorithm(ordinary_file, outlier_file, k_clusters, seg_lens, save_results=True)
    summary = sc.analyse(file_name, k_clusters, seg_lens, save_histograms=True, save_grid=True)
    valid1_summary = pd.DataFrame(summary[2,1:,1:]/1000, 
                                  index=["K"+str(param) for param in summary[2,1:,0]], 
                                  columns=["Len"+str(param) for param in summary[2,0,1:]])
    print("file_name: {}".format(file_name))
    print("F1 metric for different sets of hyperparameters. \nF1 values between 0-1, the higher the better")
    print(valid1_summary.round(2))
    return summary
```

This outlier detection algorithm was trained on a set of synthetic heartbeat/rho time series and tested for its ability to recognise the difference between a validation set of rho ("ordinary" time series) and a set of a different type of time series ("outliers"). This procedure was repeated for a number of outlier types (listed in outlier_datasets), where the main types are a sine wave, a box function and a flat time series. It was also tested on detecting the rho type after being tested on one of the outlier types.

# Outline of (1) the algorithm and (2) the analysis of its output
The algorithm uses functions from the segment_cluster.py (sc) script. The results below were generated as follows:
For each outlier type:
    For each combination of hyperparameters (number of clusters for k-means (k_clusters), and length of a time series segment (seg_len)):
        1. Run Validate_Algorithm. It creates a TSSCOD (time series segment clustering, outlier detection) class object with given values of k_clusters and seg_len.
            a) Algorithm is trained with TSSCOD.train;
                i) For every training time series, use sc.segmentation to create an exhaustive list (stride = 1) of subsegments of length seg_len.
                ii) Taking each subsegment as a vector of size seg_len, cluster them with the K-means algorithm.
            b) Perform validation with TSSCOD.validate; For every validation time series (subsets from both ordinary and outlier sets):
                i) Use sc.segmentation with a stride = seg_len, i.e. create segments with no overlap.
                ii) Reconstruct the original time series with cluster centroids that match the segments best.
                iii) Save the reconstruction error; mean squared difference between standardised values of data points of the time series and its reconstruction.
        2. Analyse the results; find what is the difference in error value between the worst rho (wr) and the best outlier (bo) reconstructions. If wr < bo then the two types of time series are completely separated, which means the rates of false positives and negatives equal to zero and the F1 metric equal to 1. If wr > bo, then a confusion zone is present, and the F1 metric is less than 1. 


```python
rho_outlier = "data/synthetic_rhos_v2.csv"
outlier_datasets = ["data/synthetic_flats.csv",
                    rho_outlier,
                    "data/synthetic_boxes.csv",
                    "data/synthetic_boxes_thick.csv",
                    "data/synthetic_boxes_thin.csv",
                    "data/synthetic_sines.csv",
                    "data/synthetic_sines_low.csv",
                    "data/synthetic_sines_long.csv",
                    "data/synthetic_sines_short.csv",
                    "data/synthetic_sines_low_long.csv",
                    "data/synthetic_sines_low_short.csv",
                   rho_outlier,
                   rho_outlier]

import os
import fnmatch
results_files=[]
wd = "/export/data/jakubok/GRS1915+105/Clusters/timeseries_segment_clustering"
for item in fnmatch.filter(os.listdir(wd), "validation_result*.csv"):
    results_files.append(item)

results_files = np.sort(np.array(results_files))

rho_data = np.loadtxt(rho_outlier, delimiter=',')
rho_outlier_counter = 0
rho_outlier_against = ["data/synthetic_flats.csv", "data/synthetic_sines.csv", "data/synthetic_boxes.csv"]
for results_ind, results in enumerate(results_files):
    summary = np.loadtxt(results.split(".")[0]+"/grid_search.csv", delimiter=',')
    outlier_data = np.loadtxt(outlier_datasets[results_ind], delimiter=',')
    if outlier_datasets[results_ind] == rho_outlier:
        outlier_data = np.loadtxt(rho_outlier_against[rho_outlier_counter], delimiter=',')
        plot_time_series(outlier_data, rho_data)
        rho_outlier_counter+=1
    else:
        plot_time_series(rho_data, outlier_data)
        
    valid_summary = pd.DataFrame(summary[1:,1:]/1000, 
                                  index=["K"+str(param) for param in summary[1:,0]], 
                                  columns=["Len"+str(param) for param in summary[0,1:]])
    print("Example training time series in blue, outliers in orange")
    print("file_names: {}\n\t\t{}".format(results, outlier_datasets[results_ind]))
    print("F1 metric for different sets of hyperparameters. \nF1 values between 0-1, the higher the better")
    print(valid_summary.round(2))
    print()
```

```python
rho_outlier = "data/synthetic_rhos_v2.csv"

# datasets used, every item corresponds to one complete test of the algorithm
outlier_datasets = ["data/synthetic_flats.csv", 
                    "data/synthetic_boxes.csv", 
                    "data/synthetic_boxes_thick.csv", 
                    "data/synthetic_boxes_thin.csv", 
                    "data/synthetic_sines.csv", 
                    "data/synthetic_sines_low.csv", 
                    "data/synthetic_sines_long.csv", 
                    "data/synthetic_sines_short.csv", 
                    "data/synthetic_sines_low_long.csv", 
                    "data/synthetic_sines_low_short.csv",
                    "data/synthetic_flats.csv", 
                    "data/synthetic_boxes.csv", 
                    "data/synthetic_sines.csv"]

# how many tests were performed on the heartbeat when trained on the outlier data; these are at the end of outlier_datasets
no_of_reverse_tests = 3


results_files=[]
wd = "/export/data/jakubok/GRS1915+105/Clusters/timeseries_segment_clustering"

# visualise results of all the tests
for item in fnmatch.filter(os.listdir(wd), "validation_result*.csv"):
    results_files.append(item)

results_files = np.sort(np.array(results_files))

rho_data = np.loadtxt(rho_outlier, delimiter=',')


for results_ind, results in enumerate(results_files[-len(outlier_datasets):-1-no_of_reverse_tests]):
    summary = np.loadtxt(results.split(".")[0]+"/grid_search.csv", delimiter=',')
    outlier_data = np.loadtxt(outlier_datasets[results_ind], delimiter=',')
    plot_time_series(rho_data, outlier_data)  
    valid_summary = pd.DataFrame(summary[1:,1:]/1000, 
                                  index=["K"+str(param) for param in summary[1:,0]], 
                                  columns=["Len"+str(param) for param in summary[0,1:]])
    print("Example training time series in blue, outliers in orange")
    print("file_names: {}\n\t\t{}".format(results, outlier_datasets[results_ind]))
    print("F1 metric for different sets of hyperparameters. \nF1 values between 0-1, the higher the better")
    print(valid_summary.round(2))
    print()
    
for results_ind, results in enumerate(results_files[-no_of_reverse_tests:-1]):
    summary = np.loadtxt(results.split(".")[0]+"/grid_search.csv", delimiter=',')
    outlier_data = np.loadtxt(outlier_datasets[results_ind], delimiter=',')
    plot_time_series(outlier_data, rho_data)  
    valid_summary = pd.DataFrame(summary[1:,1:]/1000, 
                                  index=["K"+str(param) for param in summary[1:,0]], 
                                  columns=["Len"+str(param) for param in summary[0,1:]])
    print("Example training time series in blue, outliers in orange")
    print("file_names: {}\n\t\t{}".format(results, outlier_datasets[results_ind]))
    print("F1 metric for different sets of hyperparameters. \nF1 values between 0-1, the higher the better")
    print(valid_summary.round(2))
    print()
```
